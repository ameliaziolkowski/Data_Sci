#Pie chart

```python
import matplotlib.pyplot as plt #have to run this bvok
```

```python
labels = ['Frogs', 'Dogs', 'Cats', 'Birds']

sizes = [15, 20, 45, 10]

explode = (0,0.1,0,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

ax1.axis('equal')
```

```python
labels = ['Frogs', 'Dogs', 'Cats', 'Birds']

sizes = [15, 20, 45, 10]

explode = (0,1,0,2)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

ax1.axis('equal')
```

```python
def flip_coin():                                                                #def means define a new variable, allignment is important
    coin = np.random.random() > 0.5                                             # shift, tab, bring everything to the right
    if coin == True:
      print("heads")                      
    else:
      print("tails") 
```
